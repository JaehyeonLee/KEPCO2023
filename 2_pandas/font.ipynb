```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(-100,100,50).cumsum()
data
```

```python
plt.rc('font', family='D2Coding')
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```

```python
import seaborn as sb
tips = sb.load_dataset('tips')
# sb.distplot(tips['total_bill']) version 0.14.0 에 사라짐
ax = sb.histplot(data = tips, x = 'total_bill',bins =7,kde=True)
ax.set_title('히스토그램')
```

```python
ax = sb.countplot(data=tips,x='day')
ax.set_title('count of days')
ax.set_xlabel('day')
ax.set_ylabel('빈도')
```

```python
sb.regplot(data=tips,x='total_bill',y='tip',fit_reg=False)
```

```python
sb.jointplot(data=tips,x='total_bill',y='tip',kind = "hex")
```

```python
sb.kdeplot(data=tips,x='total_bill',y='tip',fill = "True")
```

```python
sb.barplot(data=tips,x='time',y='total_bill')
```

```python
sb.boxplot(data=tips,x='sex',y='total_bill')
```

```python
sb.pairplot(tips)
```

```python
sb.violinplot(data=tips,y='total_bill',x='time',hue='sex',split=True)
```

```python
sb.lmplot(y='tip',x='total_bill',data=tips,hue='sex',fit_reg=False)
```
